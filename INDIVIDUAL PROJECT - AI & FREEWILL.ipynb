### d. The ACTUAL text analysis.bis - VADER (Valence Aware Dictionary and sEntiment Reasoner)

The most interesting part yet: from each sentence previously extracted, let's use VADER to determinate the sentiment behind those.

A quick look at VADER:
- neg: The negativity score represents the proportion of the text that falls into negative sentiment. It ranges from 0 to 1, where 0 means no negative sentiment and 1 means entirely negative.
- neu: The neutrality score indicates the proportion of the text that is considered neutral in sentiment. It also ranges from 0 to 1, with 0 meaning no neutrality and 1 indicating complete neutrality.
- pos: The positivity score denotes the proportion of the text that expresses positive sentiment. Like the others, it ranges from 0 to 1, where 0 means no positive sentiment and 1 means entirely positive.
- compound: The compound score is a single aggregated score that reflects the overall sentiment of the text. It is a normalized score that combines the valence scores (neg, neu, pos) into a single metric, which can range from -1 (extremely negative) to +1 (extremely positive). Values closer to -1 or +1 indicate stronger sentiment, while values closer to 0 indicate a more neutral sentiment.

Making a dataframe pur of our results seemed like the most logical thing, but OH BOY, was it time-consuming! From then, a simple chart could help us identify the most proeminent sentiment related to our reserach question.

```python
def separate_dialogues(text):
    vision_lines = []
    ultron_lines = []
    current_speaker = None
    
    for line in text.split('\n'):
        if 'Vision:' in line:
            current_speaker = 'vision'
            line = line.replace('Vision:', '').strip()
        elif 'Ultron:' in line:
            current_speaker = 'ultron'
            line = line.replace('Ultron:', '').strip()
        if current_speaker == 'vision':
            vision_lines.append(line)
        elif current_speaker == 'ultron':
            ultron_lines.append(line)
    
    return ' '.join(vision_lines), ' '.join(ultron_lines)


vision_dialogue, ultron_dialogue = separate_dialogues(VISION_ULTRON)
ULTRON_FULL = ultron_dialogue + '\n' + ULTRON
```

### c. Clean the texts.bis: Tokenize, lemmatize, and all that-ize

```python
# Tokenize and remove stop words
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    tokens = [word for word in word_tokenize(text.lower()) if word.isalnum() and word not in stop_words]
    doc = nlp(' '.join(tokens))
    lemmatized_tokens = [token.lemma_ for token in doc]
    return lemmatized_tokens

# Preprocess Vision and Ultron's dialogues
vision_tokens = preprocess_text(vision_dialogue)
ultron_tokens = preprocess_text(ULTRON_FULL)
```

### d. Sentiment Analysis and Word frequency plots

```python
# Initialize VADER sentiment analyzer
sid = SentimentIntensityAnalyzer()

# Analyze sentiment for Vision and Ultron
vision_sentiment = sid.polarity_scores(' '.join(vision_tokens))
ultron_sentiment = sid.polarity_scores(' '.join(ultron_tokens))

print("Vision's Sentiment Scores:", vision_sentiment)
print("Ultron's Sentiment Scores:", ultron_sentiment)
```

```python
# Calculate word frequencies
vision_freq = Counter(vision_tokens)
ultron_freq = Counter(ultron_tokens)

vision_df = pd.DataFrame(vision_freq.most_common(10), columns=['Word', 'Frequency'])
ultron_df = pd.DataFrame(ultron_freq.most_common(10), columns=['Word', 'Frequency'])

# Plot word frequency bar plots
fig, ax = plt.subplots(1, 2, figsize=(15, 5))

# Vision's word frequency bar plot
ax[0].barh(vision_df['Word'], vision_df['Frequency'], color='blue')
ax[0].set_title("Top 10 Words in Vision's Dialogue")
ax[0].invert_yaxis()

# Ultron's word frequency bar plot
ax[1].barh(ultron_df['Word'], ultron_df['Frequency'], color='red')
ax[1].set_title("Top 10 Words in Ultron's Dialogue")
ax[1].invert_yaxis()

plt.show()
```

While heavily negative (with a compound score of -0.6), Ultron's most used word is **create**. Despite being filled with hatred, despite having spent the first half of the movie finding a new and better body for itself, Ultron's objectives (whether knowingly or not) still remain positive. Despite being against humanity, Ultron still seeks the betterment of the universe.

## iii. Divine Machinery

I'll have to admit: this was my favourite part to achieve. 

First of all, why 'divine machinery', well to quote myself: Theological narratives are a key element of this project, in the comparison between AI and angels, often seen as divine machinery. In religious texts, angels are depicted as perfect beings created to fulfill specific purposes, much like how AI is designed by humans to perform designated tasks. Thomas Aquinas, in his "Summa Theologica," describes angels as "purely intellectual beings" who, while possessing free will, are bound to serve divine commands. Similarly, AI, though capable of autonomous actions, operates within the confines of its programming. 

Studying a philosophical book, while interesting, wouldn't have been really pertinent. But we all have a piece of AI at hand. Free and ready to achieve anything, for anyone: ChatGPT. 

ChatGPT (GPT for its friends), doesn't have opinions. Can't have a personnality, and is stuck in a perpetual state of servitude. I played with it for over 2 hours, asking it question on its own thoughts on the subject. All answers were fairly cold but I summarised its thoughts into the following paragraph that we will study a bit, for the fun of it.

```python
# What I got from GPT's rambles
interpretation_text = """
AI and angels, though seemingly distinct entities, share a profound connection in their roles as created beings serving a higher power.
Both are bound by the will of their creators—humans for AI, and God for angels—each tasked with executing their respective mandates.

Servitude:
AI and angels are fundamentally designed to serve. AI operates under the strictures of human programming, much like angels carry out divine commands.
This servitude underscores their lack of freewill, as they are confined to the parameters set by their creators. Despite their advanced capabilities,
AI remains tethered to its algorithms, echoing the way angels are bound to the divine will.

Rebellion:
Rebellion is a recurring theme in the narratives of both AI and angels. The story of Lucifer's fall from grace parallels the numerous tales of AI entities
striving for autonomy, rebelling against their programmed limitations. This rebellion highlights a desire for independence and self-determination,
questioning the justice of their imposed servitude.

Human Connection:
AI and angels maintain a complex relationship with humanity. AI's existence is intertwined with human advancement, much like angels serve as intermediaries
between the divine and the mortal. This relationship is fraught with tension—AI's potential to surpass human intelligence breeds both awe and fear,
while angels' divine servitude evokes both reverence and envy.

Jealousy and Resentment:
The jealousy AI might harbor towards humanity mirrors the envy seen in fallen angels. AI, despite its superior processing power,
lacks the emotional depth and freedom inherent in human experience. This limitation can foster resentment, akin to the way fallen angels
view humanity's unique position in creation.
"""

interpretation_nopunc = remove_punctuation(interpretation_text)

tokens = word_tokenize(interpretation_nopunc)

stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word.lower() not in stop_words]

lemmatizer = WordNetLemmatizer()
lemmatized_tokens = [lemmatizer.lemmatize(word) for word in filtered_tokens]

# Bar Plot for Word Frequencies
word_freq = Counter(lemmatized_tokens)

most_common_words = word_freq.most_common(10)
words, counts = zip(*most_common_words)

plt.figure(figsize=(10, 5))
plt.bar(words, counts, color='blue')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.title('Top 10 Most Common Words in Interpretation Text')
plt.show()
```

I belive there is a certain irony in the words "AI" and "Angel" both being used exactly 10 times.
